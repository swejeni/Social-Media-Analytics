### SMA Lab Assignment 1: Exploring networks using networkx package 

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
print(G.nodes()) 
print(G.edges())
```

```python
G.add_node("A")
```

```python
G.add_nodes_from(["B","C","D","E"])
```

```python
G.add_edge(*("A","B"))
```

```python
G.add_edges_from([("A","C"), ("B","D"), ("B","E"), ("C", "E")])
```

#### Accessing vertex and edge sets

```python
print("Vertex set: ",G.nodes())
```

```python
print("Edge set: ",G.edges())
```

#### Drawing graph

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
plt.show() 
```

```python
plt.savefig("graph.png") 
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

#### Adjacency view

```python
print(G.adj)
```

#### Degree of a vertex

```python
G.degree("A")
```

#### Creating weighted graph

```python
G = nx.Graph() 
E = [('A', 'B', 2), ('A', 'C', 1), ('B', 'D', 5), ('B', 'E', 3), ('C', 'E', 2)]
G.add_weighted_edges_from(E)
```

```python
pos=nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold')
edge_weight = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight)
plt.show()
```
